```python
import random 

print("Rules for the RPS game : \n"
        +"ROCK vs PAPER->PAPER wins the game \n"
        + "ROCK vs SCISSOR->ROCK wins the game \n"
        +"PAPER vs SCISSOR->SCISSOR wins the game \n") 

while True: 
    print("Enter choice \n 1. ROCK \n 2. PAPER \n 3. SCISSOR \n") 

    ch = int(input("player turn: ")) 

    while ch > 3 or ch< 1: 
        ch = int(input("enter valid input: ")) 

    if ch == 1: 
        ch_name = 'ROCK'
    elif ch == 2: 
        ch_name = 'PAPER'
    else: 
        ch_name = 'SCISSOR'
    print("then player choice is: " + ch_name) 
    print("\n its  now the turn of computer turn") 

    comp_ch = random.randint(1, 3) 

    while comp_ch == ch: 
        comp_ch = random.randint(1, 3) 

    if comp_ch== 1: 
        comp_ch_name = 'ROCK'
    elif comp_ch == 2: 
        comp_ch_name = 'PAPER'
    else: 
        comp_ch_name = 'SCISSOR'
        
    print("then the Computer choice is: " + comp_ch_name) 

    print(ch_name + " V/s " + comp_ch_name) 

    # condition for win
    if((ch == 1 and comp_ch == 2) or
        (ch == 2 and comp_ch ==1 )): 
        print("PAPER wins => ", end = "") 
        result = "PAPER"
        
    elif((ch == 1 and comp_ch == 3) or
        (ch == 3 and comp_ch == 1)): 
        print("ROCK wins =>", end = "") 
        result = "ROCK"
    else: 
        print("Scissor wins =>", end = "") 
        result = "SCISSOR"

    # Printing either user or computer wins 
    if result == ch_name: 
        print("<== PLAYER wins ==>") 
    else: 
        print("<== COMPUTER wins ==>") 
        
    print("Do you want to play again? (Y/N)") 
    ans = input() 


    # if user input n or N then condition is True 
    if ans == 'n' or ans == 'N': 
        break

print("\nThanks ")

```
